```python
import cv2
import math
import numpy as np
import os

from deepface import DeepFace
from deepface.commons import functions
from matplotlib import pyplot as plt
from pathlib import Path
```

```python
cwd = os.getcwd()
base_img_path = cwd+'/../data/IM_'
plt.rcParams['figure.figsize'] = [60, 30]
```

```python
def nlm_denoise_edge_detection(img):
    b,g,r = cv2.split(img)           # get b,g,r
    rgb_img = cv2.merge([r,g,b])     # switch it to rgb

    # Denoising
    nlm_img = cv2.fastNlMeansDenoisingColored(img,None,10,10,7,21)
    b,g,r = cv2.split(nlm_img)           # get b,g,r
    rgb_nlm_img = cv2.merge([r,g,b])     # switch it to rgb
    
    edges = cv2.Canny(img,None,100,200,3,True)
    nlm_edges = cv2.Canny(nlm_img,None,100,200,3,True)
    
    return edges, nlm_edges
```

```python
def locate_face(face, image):
    x,y,w,h = 0,0,face.shape[1],face.shape[0]
    dist_min = -1
    face_b, face_g, face_r = cv2.split(face)
    # Iterate through all windows to find the one closest to the extracted face
    for i in range(image.shape[0]-h): # row
        for j in range(image.shape[1]-w): # column
            window = image[i:i+h, j:j+w]
            window_b,window_g,window_r = cv2.split(window)
            dist = np.linalg.norm(window_b-face_b)+np.linalg.norm(window_g-face_g)+np.linalg.norm(window_r-face_r)
            if(dist_min<0 or dist<dist_min):
                x = j
                y = i
                dist_min = dist
                # print(f'found new dist = {dist} at {x},{y}')
    return x,y,w,h,dist_min
```

```python
def detect_face_location(img):
    face_raw = functions.detect_face(img, 'dlib')
    plt.imshow(face_raw[:,:,::-1])
    x,y,w,h,dist = locate_face(face_raw, img)
    return x,y,w,h,face_raw
```

```python
def build_mask(img,x,y,w,h):
    print(f'x={x},y={y},w={w},h={h}')
    face_mask = np.zeros(img.shape[:2])
    for i in range(y,y+h+1):
        for j in range(x,x+w+1):
            face_mask[i][j] = 1
    return face_mask
```

```python
def compute_filter(window_edge, window_body, edge_weight):
    # Put 1 anywhere that is part of the body, and <edge_weight> in non-body area that has an edge detected
    # <edge_weight> between 0 and 1
    assert edge_weight>=0 and edge_weight<=1, "edge_weight must be between 0 and 1 inclusively."
    filt = np.zeros(window_edge.shape[:2])
    for i in range(filt.shape[0]):
        for j in range(filt.shape[1]):
            if(window_body[i][j]==1):
                filt[i][j] = 1
            elif(window_edge[i][j]==255):
                filt[i][j] = edge_weight
    return filt
```

```python
def pre_process_edge_window(window_edge, window_body):
    # Place 1 anywhere that is a body area or has an edge
    post_process = np.zeros(window_edge.shape[:2])
    for i in range(post_process.shape[0]):
        for j in range(post_process.shape[1]):
            if(window_body[i][j]==1 or window_edge[i][j]==255):
                post_process[i][j] = 1
    return post_process
```

```python
def weighted_average(window, filt):
    total_val = 0
    total_weight = 0
    # print(f'window = {window},\n weight = {weights}')
    for i in range(window.shape[0]):
        for j in range(window.shape[1]):
            if(filt[i][j]==1):
                total_val += filt[i][j]
            else:
                total_val += window[i][j]*filt[i][j]
            total_weight += 1
    return total_val/total_weight
```

```python
def apply_filter(edge, body_mask, row, col, width, height, weight):
    # w = 5, h = 10
    # r = 6, c = 20
    # i = range(4,8), j = range(11,21)
    if(body_mask[row][col]==1):
        return 1
    first_col = max(0,col-int(width/2))
    last_col = min(edge.shape[1]-1, col+int(width/2))
    first_row = max(0,row-int(height/2))
    last_row = min(edge.shape[0]-1, row+int(height/2))
    # print(f'col = {col}, width = {width}, first_col = {first_col}, last_col = {last_col}')
    window_edge = edge[first_row:last_row, first_col:last_col+1] # inclusive range
    window_body = body_mask[first_row:last_row, first_col:last_col+1] # inclusive range
    filt = compute_filter(window_edge, window_body, weight)
    window_edge_processed = pre_process_edge_window(window_edge, window_body)
    # print(f'filt = {filt}')
    # print(f'window_edge_processed = {window_edge_processed}\n')
    return weighted_average(window_edge_processed, filt)
```

```python
def segment_body(edge,body_mask,i,j,window_w,window_h,weight,threshold):
    filtered_val = apply_filter(edge, body_mask, i, j, window_w, window_h, weight)
    # print(f'({i},{j},{filtered_val})')
    return 1 if filtered_val > threshold else 0
```

```python
image_idx = '04'
img_path = base_img_path + image_idx +'.jpg'
```

```python
img = cv2.imread(img_path)
edge, nlm_edge = nlm_denoise_edge_detection(img)

fig, axes = plt.subplots(1,2,figsize=(20,80))
axes[0].imshow(edge,cmap = 'gray')
axes[1].imshow(nlm_edge,cmap = 'gray')
```

```python
try:
    x,y,w,h,face_raw = detect_face_location(img)
except ValueError as e:
    print(f'Face detection failed: '+str(e))
```

```python
mask = build_mask(img,x,y,w,h)
```

```python
body_mask = mask.copy()
```

```python
row_start = y
col_start = x+int(w/2)
body_face_ratio = 4
window_face_ratio = 0.4
top_face_ratio = 0.6
# horizontal max width = ratio*w = 516
```

```python
window_w = int(w*window_face_ratio)
window_h = int(h*window_face_ratio)
weight = 0.5
threshold = 0.5
top_row = y - int(h*top_face_ratio)
```

```python
output = np.zeros(img.shape[:2])
for dist_row in range(img.shape[0]):
    # go up
    i = row_start-dist_row
    if(i>=top_row):
        print(f'Processing row {i}')
        filtered_row = np.zeros(body_mask[i].shape[:1])
        for dist_col in range(int(w*body_face_ratio/2)):
            j_left = col_start-dist_col
            j_right = col_start+dist_col
            output[i][j_left] = segment_body(nlm_edge,body_mask,i,j_left,window_w,window_h,weight,threshold)
            filtered_row[j_left] = output[i][j_left]
            output[i][j_right] = segment_body(nlm_edge,body_mask,i,j_right,window_w,window_h,weight,threshold)
            filtered_row[j_right] = output[i][j_right]
            # print(f'j_left = {j_left}, j_right = {j_right}')
            if(output[i][j_left]==1):
                print(f'Detected at ({j_left},{i})')
            if(output[i][j_right]==1):
                print(f'Detected at ({j_right},{i})')
        body_mask[i] = filtered_row
    # go down
    i = row_start+dist_row
    if(i<img.shape[0]):
        print(f'Processing row {i}')
        filtered_row = np.zeros(body_mask[i].shape[:1])
        for dist_col in range(int(w*body_face_ratio/2)):
            j_left = col_start-dist_col
            j_right = col_start+dist_col
            output[i][j_left] = segment_body(nlm_edge,body_mask,i,j_left,window_w,window_h,weight,threshold)
            filtered_row[j_left] = output[i][j_left]
            output[i][j_right] = segment_body(nlm_edge,body_mask,i,j_right,window_w,window_h,weight,threshold)
            filtered_row[j_right] = output[i][j_right]
            # print(f'j_left = {j_left}, j_right = {j_right}')
            if(output[i][j_left]==1):
                print(f'Detected at ({j_left},{i})')
            if(output[i][j_right]==1):
                print(f'Detected at ({j_right},{i})')
        body_mask[i] = filtered_row
        
```

```python
edge_result = nlm_edge.copy()
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if output[i][j]==0:
            edge_result[i][j] = 0
            
plt.imshow(edge_result,cmap = 'gray')
```

```python
save_result_path = base_img_path + image_idx +'_edge_v5_'+str(window_w)+'_'+str(window_h)+'_'+str(weight)+'_'+str(threshold)+'_.jpg'
cv2.imwrite(save_result_path, edge_result)
```
